**BOXPLOT / WHISKER PLOT**

```python
sns.boxplot(df1['N2O'],color='Red')
Q3=df.N2O.quantile(0.75)
Q1=df.N2O.quantile(0.25)
IQR=Q3-Q1
Median=df.N2O.median()
print("Q1 Value:",Q1)
print("Median Value:",df.N2O.median())
print("Q3 Value:",Q3)
print("Upper whisker limit:",(Q3 + 1.5*IQR))
print("Lower whisker limit:",(Q1 - 1.5*IQR))
```

```python
sns.boxplot(df1['CO2'],color='Pink')
Q3=df.CO2.quantile(0.75)
Q1=df.CO2.quantile(0.25)
IQR=Q3-Q1
Median=df.CO2.median()
print("Q1 Value:",Q1)
print("Median Value:",df.CO2.median())
print("Q3 Value:",Q3)
print("Upper whisker limit:",(Q3 + 1.5*IQR))
print("Lower whisker limit:",(Q1 - 1.5*IQR))
```

```python
sns.boxplot(df1['TSI'],color='Purple')
Q3=df.TSI.quantile(0.75)
Q1=df.TSI.quantile(0.25)
IQR=Q3-Q1
Median=df.TSI.median()
print("Q1 Value:",Q1)
print("Median Value:",df.TSI.median())
print("Q3 Value:",Q3)
print("Upper whisker limit:",(Q3 + 1.5*IQR))
print("Lower whisker limit:",(Q1 - 1.5*IQR))
```
